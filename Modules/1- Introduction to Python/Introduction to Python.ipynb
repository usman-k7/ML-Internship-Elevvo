**3) Variables:**

What are Variables?

📌	Variables are containers that store numbers, strings, and other data types and structures.

📌	Variables are names given to values that can be changed.

📌	Variables are assigned values using the equal sign (=).


a) Creating Variables:

```python
x = 'Hello'
name = "Ahmed"
y = 10
```

```python
print(x)
print(name)
print(y)
```

```python
#or we can just type the variable to print it
x
```

📌 we cannot print a variable that we did not declared

*for example*: print(z) will raise error:  *NameError: name 'z' is not defined*

```python
x = 4
print(x)   # x is of type int
x = "Sally" # x is now of type str
print(x)
```

b) Casting:

📌 If you want to specify the data type of a variable, this can be done with casting.

```python
x = str(3)    # x will be '3'
y = int(x)    # y will be 3
z = float(3)  # z will be 3.0
```

```python
print(x)
print(y)
print(z)
```

c) Get the Type:

📌 You can get the data type of a variable with the type() function.

```python
x = 5
y = "John"
print(type(x))
print(type(y))
```

e) Variable Names:

📌 A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume).

📌 Rules for Python variables:

1. A variable name must start with a letter or the underscore character.

2. A variable name cannot start with a number.

3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ).

4. Variable names are case-sensitive (age, Age and AGE are three different variables).



```python
# Correct Variable names
myvar = "John"
my_var = "John"
_my_var = "John"
myVar = "John"
MYVAR = "John"
myvar2 = "John"
```

📌 incorrect variables names like starting with a digit, including '-' or spaces at the name, will raise an error: *SyntaxError: invalid decimal literal*

**6) Python Tuples:**

**7) Python Sets:**
